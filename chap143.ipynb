##### 1.4.3 Geometric Transformations of Images ####

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

##### scale #####

```python
#이미지 크기(스케일) 조정
img = cv2.imread('Hero.png')
res = cv2.resize(img,None,fx=3, fy=3, interpolation = cv2.INTER_CUBIC)
#cv2.INTER_LINEAR
```

```python
cv2.namedWindow('image')
while(1):
    cv2.imshow('image',res)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
```

##### translate #####

```python
#x축으로100,y축으로50 이동
#M은 변환행렬
img = cv2.imread('Hero.png')
rows,cols = img.shape[:2]
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### rotate #####

```python
#cols/2, rows/2라서 중심 기준으로 제일 뒤에서 두번째 값만큼 각도를 돌림
#M은 변환행렬
img = cv2.imread('Hero.png')
rows,cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Affine Transformation

```python
#BGR2RGB? 하면 제대로 색깔 출력 가능
img = cv2.imread('Hero.png')
img2=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img2,M,(cols,rows))
plt.subplot(121),plt.imshow(img2),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

##### Perspective Transformation

```python
img = cv2.imread('Hero2.png')
img2=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows,cols,ch = img.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img2,M,(300,300))
plt.subplot(121),plt.imshow(img2),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
