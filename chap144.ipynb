#### 1.4.4 Smoothing Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Hero.png')
#5x5사이즈로 컨볼루션 연산하고 25로 나눠서 연하게
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
#-1은 출력 이미지의 깊이로, -1은 입력 이미지와 동일한 깊이를 의미함
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

##### 1. Averaging, Gaussian Filtering, Median Filtering, Bilateral Filtering

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

##### 1. Averaging

```python
#큰숫자 넣으면 정도 심해짐
img = cv2.imread('Hero.png')
blur1 = cv2.blur(img,(3,3))
blur2 = cv2.blur(img,(10,10))
plt.subplot(121),plt.imshow(blur1),plt.title('blur1')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur2),plt.title('blur2')
plt.xticks([]), plt.yticks([])
plt.show()
#blur = cv2.GaussianBlur(img,(5,5),0)
#median = cv2.medianBlur(img,5)
#blur = cv2.bilateralFilter(img,9,75,75)
```

##### 2. Gaussian Filtering

```python
img = cv2.imread('Hero.png')
blur3 = cv2.GaussianBlur(img,(7,7),0)
plt.subplot(121),plt.imshow(img),plt.title('Hero.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur3),plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.show()
```

##### 3. Median Filtering

```python
img = cv2.imread('Hero.png')
median = cv2.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Hero.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Median')
plt.xticks([]), plt.yticks([])
plt.show()
```

##### 4. Bilateral Filtering

```python
img = cv2.imread('Hero.png')
blur4 = cv2.bilateralFilter(img,9,75,75)
plt.subplot(121),plt.imshow(img),plt.title('Hero.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur4),plt.title('Bilateral')
plt.xticks([]), plt.yticks([])
plt.show()
```
