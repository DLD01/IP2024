#### 1.4.5 Morphological Transformations

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

##### 1. Erosion

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray'),plt.title('j.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(erosion, cmap='gray'),plt.title('Erosion')
plt.xticks([]), plt.yticks([])
```

##### 2. Dilation

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray'),plt.title('j.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dilation, cmap='gray'),plt.title('Dilation')
plt.xticks([]), plt.yticks([])
```

##### 3. Opening

```python
img = cv2.imread('j2.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray'),plt.title('j2.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(opening, cmap='gray'),plt.title('Opening')
plt.xticks([]), plt.yticks([])
```

##### 4. Closing

```python
img = cv2.imread('j3.png',0)
kernel = np.ones((7,7),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray'),plt.title('j3.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(closing, cmap='gray'),plt.title('Closing')
plt.xticks([]), plt.yticks([])
```

##### 5. Morphological Gradient

```python
img = cv2.imread('j.png',0)
kernel = np.ones((5,5),np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray'),plt.title('j.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(gradient, cmap='gray'),plt.title('Morphological Gradient')
plt.xticks([]), plt.yticks([])
```

##### 6. Top Hat 원본에서 Opening을 뺀 결과

```python
img = cv2.imread('j2.png',0)
kernel = np.ones((13,13),np.uint8)#5대신13
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray'),plt.title('j2.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(tophat, cmap='gray'),plt.title('Top Hat')
plt.xticks([]), plt.yticks([])
```

##### 7. Black Hat Closing에서 원본을 뺀 결과

```python
img = cv2.imread('j.png',0)
kernel = np.ones((9,9),np.uint8)#5대신20
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray'),plt.title('j.png')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blackhat, cmap='gray'),plt.title('Black Hat')
plt.xticks([]), plt.yticks([])
```
